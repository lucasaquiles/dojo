```python
import numpy as np

def sigmoid(soma):
    return 1 / (1 + np.exp(-soma))

def sigmoidDerivada(sig):
    return sig * (1 - sig)

entradas = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

saidaEsperada = np.array([
    [0],[1],[1],[0]
])

pesos0 = np.array([
    [-0.424, -0.740, -0.961],
    [0.358, -0.577, -0.469],
])

pesos1 = np.array([
    [-0.017],
    [-0.893],
    [0.148]
])

epocas = 100

#dot faz o x*w + x2*w2
for j in range(epocas):

    camadaEntrada = entradas
    somaSinapse0 = np.dot(entradas, pesos0)

    #print(somaSinapse0)
    camadaOculta = sigmoid(somaSinapse0)

  #  print(camadaOculta)

    camadaFinal = np.dot(camadaOculta, pesos1)

   # print(camadaFinal)
    camadaSaida = sigmoid(camadaFinal)
    #print(camadaSaida)
    
    errosCamadaSaida = saidaEsperada- camadaSaida
    print(errosCamadaSaida)
    mediaAbsoluta = np.mean(np.abs(errosCamadaSaida))
    print("média absoluta: "+str(mediaAbsoluta))

    
#subtrair a saida gerada pela saída esperada
# erro = respostaCorreta - respostaCalculada
# outra forma (elevação do erro ao quadrado)
# media de erros (soma os valores / quantidade de registros)
```
